# Sample Prices

```python
from mplchart.samples import sample_prices, SAMPLE_FREQUENCIES
```

```python
for freq in SAMPLE_FREQUENCIES:
    prices = sample_prices(freq)
    index = prices.index
    print(freq, len(prices), index.dtype)
    
```

```python
for freq in SAMPLE_FREQUENCIES:
    prices = sample_prices(freq, backend="polars")
    print(freq, len(prices), prices['datetime'].dtype)

```
