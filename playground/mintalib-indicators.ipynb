# Mintalib Indicators

Using [mintalib](https://pypi.org/project/mintalib/) indicators with `mplchart` on the pandas backend.
`mintalib.indicators` follows the same calling convention as `mplchart.indicators` and plugs directly into `Chart.plot`.

```python
from mplchart.chart import Chart
from mplchart.samples import sample_prices
from mplchart.primitives import Candlesticks, Volume, LinePlot
from mintalib.indicators import SMA, RSI, MACD, BBANDS, DMI

prices = sample_prices()
```

## SMA

Simple Moving Average overlay.

```python
Chart(prices, title="SMA", max_bars=250).plot(
    Candlesticks(),
    SMA(20),
    SMA(50),
).show()
```

## RSI

Relative Strength Index with overbought / oversold lines.

```python
Chart(prices, title="RSI", max_bars=250).plot(
    Candlesticks(), SMA(50), Volume(),
).pane("above", yticks=(30, 50, 70)).plot(
    RSI(14) | LinePlot(overbought=70, oversold=30),
).show()
```

## MACD

Moving Average Convergence Divergence.

```python
Chart(prices, title="MACD", max_bars=250).plot(
    Candlesticks(), SMA(50), Volume(),
).pane("below").plot(
    MACD(),
).show()
```

## Bollinger Bands

```python
Chart(prices, title="Bollinger Bands", max_bars=250).plot(
    Candlesticks(), Volume(), BBANDS(),
).show()
```

## DMI

Directional Movement Index: ADX with +DI and -DI.

```python
Chart(prices, title="DMI", max_bars=250).plot(
    Candlesticks(), SMA(50), Volume(),
).pane("below", yticks=(20, 40)).plot(
    DMI(14),
).show()
```
