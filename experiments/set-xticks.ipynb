```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data with missing dates (e.g., weekends)
dates = pd.to_datetime(['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-08', '2024-01-09'])
values = [10, 12, 15, 18, 20] # Note the gap between Jan 3rd and Jan 8th

# Create an integer index (0, 1, 2, 3, 4) for the x-axis
x_indices = np.arange(len(dates))

plt.figure(figsize=(8, 5))
plt.plot(x_indices, values, marker='o')

# Set the x-axis tick locations to the integer indices
plt.xticks(x_indices, [date.strftime('%Y-%m-%d') for date in dates], rotation=20)
# Add ha = "right" to align the labels to the right

plt.xlabel('Date (Gaps Removed)')
plt.ylabel('Value')
plt.title('Plotting Time Series without Weekend Gaps')
plt.grid(True)
plt.tight_layout() # Helps prevent labels from overlapping
plt.show()
```
