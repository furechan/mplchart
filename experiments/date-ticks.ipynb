```python

def interval_freq(interval):
    interval = interval / np.timedelta64(1, 's')

    if interval < 60:
        return 's'
    elif interval < 3600:
        return 'm'
    elif interval < 3600 * 24:
        return 'h'
    elif interval < 3600 * 24 * 7:
        return 'D'
    elif interval < 3600 * 24 * 30:
        return 'W'
    elif interval < 3600 * 24 * 365:
        return 'M'
    else:
        return 'Y'


def date_ticks(dates, count=10):
    dates = np.asarray(dates, 'datetime64[s]')

    if count <= 0:
        return []

    if len(dates) <= count:
        return np.arange(len(dates))

    interval = (dates[-1] - dates[0]) / count
    freq = interval_freq(interval)

    values = dates.astype(f"datetime64[{freq}]").astype(int)

    step = (values[-1] - values[0]) / count
    values = (values / step).astype(int)

    print(interval, freq, step)

    mask = np.concatenate(([False], values[1:] != values[:-1]))

    return np.where(mask)[0]

date_ticks(dates, 1)

```
