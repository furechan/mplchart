```python
import logging

import numpy as np
import pandas as pd

logging.basicConfig(level=logging.DEBUG)
```

```python
dates = np.array([
    "2024-01-05",
    "2024-01-20",
    "2024-02-03",
    "2024-03-15",
    "2024-03-28",
    "2024-04-20",
    "2024-05-20",
    "2024-06-20",
    ], dtype="datetime64[D]")

"""
Pandas Key Frequencies (freq):
'D': Daily
'W': Weekly
'M' / 'MS': Month End / Month Start
'Q' / 'QS': Quarter End / Quarter Start
'Y' / 'YS': Year End / Year Start 
 """


dates = pd.date_range(start="1999-12-31", periods=100, freq="D").to_numpy()
dates
```

```python

# Extract month, find where it changes
values = dates.astype("datetime64[W]").astype(int) // 1
mask = np.concatenate(([False], values[1:] != values[:-1]))  # include first date or not, up to you

dates[mask]
# array(['2024-01-05', '2024-02-03', '2024-03-15'], dtype='datetime64[D]')
```

```python
months = dates.astype("datetime64[M]").astype(int) // 2
mask = np.concatenate(([True], months[1:] != months[:-1]))  # include first date or not, up to you
mask = np.concatenate(([False], months[1:] != months[:-1]))  # include first date or not, up to you

dates[mask]
```

```python
indices = np.where(mask)[0]
indices
```

```python
from mplchart.datetimes import date_ticks

mask = date_ticks(dates, 10)

dates[mask]

```
