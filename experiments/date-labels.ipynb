```python
from mplchart.samples import sample_prices
from mplchart.datetimes import date_ticks, date_labels

prices = sample_prices(freq="daily")
prices
```

```python
dates = prices.index.tz_localize(None)
dates
```

```python
import matplotlib.pyplot as plt

def plot_prices(prices, max_bars, title=None):
    prices = prices.tail(max_bars)
    dates = prices.index.tz_localize(None)

    fig, ax = plt.subplots(figsize=(10, 1))

    if title is not None:
        ax.set_title(title)

    ax.plot(prices.close.values)
    ax.set_xmargin(0)

    ticks = date_ticks(dates, count=10)
    labels = date_labels(dates[ticks])

    ax.set_xticks(ticks, labels=labels)


prices = sample_prices(freq="minute")
plot_prices(prices, max_bars=1000)
```

```python
freq = "daily"
prices = sample_prices(freq=freq)

for max_bars in [10, 20, 50, 100, 250, 500, 1000, 2000, 5000]:
    title = f"{freq} - {max_bars}"
    plot_prices(prices, max_bars=max_bars, title=title)
```

```python
freq = "hourly"
prices = sample_prices(freq=freq)

for max_bars in [10, 20, 50, 100, 250, 500, 1000, 2000, 5000]:
    title = f"{freq} - {max_bars}"
    plot_prices(prices, max_bars=max_bars, title=title)
```
