## Visualize local peaks and valleys with Peaks

The Peaks primitive will by default apply to the previously rendered indicator

```python
import yfinance as yf

from mplchart.chart import Chart

from mplchart.primitives import Candlesticks, OHLC, Price, Volume, Peaks

from mplchart.indicators import SMA, EMA, ROC, MACD, PPO, SLOPE
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
chart = Chart(prices, title=ticker, max_bars=250, figsize=(12, 9)).plot(
    Candlesticks(),
).plot(
    SLOPE(20),
    Peaks(5),
    target="below",
).show()
```
