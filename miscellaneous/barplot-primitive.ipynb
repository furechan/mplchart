# BarPlot Primitive

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Price, Volume, BarPlot
from mplchart.indicators import SMA, EMA, RSI, ROC, ATRP, MACD, DMI
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
Chart(prices, title=ticker, max_bars=50).plot(
    Candlesticks(),
).plot(
    ROC(20) | BarPlot(color="red", alpha=0.5, width=0.8), target = "below"
).show()
```
