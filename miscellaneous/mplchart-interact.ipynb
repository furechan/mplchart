# Using mplchart interactively


```python
import matplotlib.pyplot as plt
from matplotlib_inline.backend_inline import set_matplotlib_formats

set_matplotlib_formats("svg")

plt.isinteractive()
```

```python
import yfinance as yf

from functools import lru_cache

@lru_cache(maxsize=32)
def fetch_prices(ticker, period="5y"):
    return yf.Ticker(ticker).history(period=period)
```

```python
from ipywidgets import widgets, interact
from mplchart.chart import Chart
from mplchart.primitives import Volume, Candlesticks
from mplchart.indicators import SMA, RSI, MACD, PPO, SLOPE


ticker = widgets.Text(
    value="aapl",
    description="Ticker:",
    continuous_update=False,
)

@interact(ticker=ticker)
def draw_chart(ticker):
    if not ticker:
        return "Nothing to show!"
    prices = fetch_prices(ticker)
    chart = Chart(prices, title=ticker.upper(), max_bars=250)
    chart.plot(Candlesticks(), SMA(50), SMA(200), Volume()).show()
```
