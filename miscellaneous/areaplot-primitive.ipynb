# AreaPlot Primitive

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Price, Volume, AreaPlot
from mplchart.indicators import SMA, EMA, RSI, ROC, ATRP, MACD, DMI
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
Chart(prices, title=ticker, max_bars=250).plot(
    Candlesticks(),
).plot(
    ROC(20) | AreaPlot(color="red", alpha=0.5), target="below"
).show()
```
