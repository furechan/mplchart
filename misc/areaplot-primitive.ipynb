# AreaPlot Primitive

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Price, Volume, AreaPlot
from mplchart.indicators import SMA, EMA, RSI, ROC, ATRP, MACD, DMI
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
max_bars = 250

indicators = [
    Candlesticks(),
    ROC(20) | AreaPlot(color="red", alpha=0.5)
]



chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```
