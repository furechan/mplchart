# Price Chart Example

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Price
from mplchart.indicators import SMA, EMA

```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
# Open High Low Close Bar Chart with Moving Average

max_bars = 250

indicators = [ Price(width=2), SMA(20), SMA(50) ]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```
