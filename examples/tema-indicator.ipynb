# Triple Exponential Moving Average Example

```python
from mplchart.chart import Chart

from mplchart.samples import sample_prices
from mplchart.primitives import Candlesticks, Volume, SameAxes
from mplchart.indicators import ROC, SMA, EMA, ADX, RSI, MACD, SLOPE, DEMA, TEMA
```

```python
# Fetch sample prices

prices = sample_prices()
prices.info()
```

```python
# Create chart with some indicators

max_bars = 250
indicators = [Candlesticks(), TEMA(20), ROC(1) @ TEMA(20)]

chart = Chart(title="TEMA Example", max_bars=max_bars)
chart.plot(prices, indicators)
```
