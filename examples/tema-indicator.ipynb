# Triple Exponential Moving Average Example

```python
from mplchart.chart import Chart

from mplchart.samples import sample_prices
from mplchart.primitives import Candlesticks, Volume
from mplchart.indicators import ROC, SMA, EMA, ADX, RSI, MACD, SLOPE, DEMA, TEMA
```

```python
# Fetch sample prices

prices = sample_prices()
prices.info()
```

```python
# Create chart with some indicators


chart = Chart(prices, title="TEMA Example", max_bars=250).plot(
    Candlesticks(), TEMA(20)
).plot(
    ROC(1) @ TEMA(20), target = "below"
).show()
```
