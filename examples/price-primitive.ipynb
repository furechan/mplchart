# Price Primitive

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Price, SameAxes
from mplchart.indicators import EMA, SMA
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
# Price plot

max_bars = 250

indicators = [ Price(color="red") ]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```

```python
# Price plot (with color)

max_bars = 250

indicators = [ Price("hlcc"),
              EMA() @ Price("hlcc") | SameAxes()]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```
