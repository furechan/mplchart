# Volume Chart Example

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume
from mplchart.indicators import SMA, EMA
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
# Candlesticks with Moving Averages and Volume

chart = Chart(prices, title=ticker, max_bars=250).plot(
    Candlesticks(),
    Volume(sma=50),
    SMA(20),
    SMA(50),
).show()
```
