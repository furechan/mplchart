# Markers Primitive Example

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume, Markers

from mplchart.indicators import SMA, RSI, MACD
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
# MArkers can be used to mark sign changes in an indicator
# By default the primitive is applied to the last indicator that was plotted

max_bars = 250

indicators = [
    Candlesticks(),
    SMA(50),
    SMA(200),
    Volume(),
    RSI(),
    MACD(),
    Markers("macdhist", color=["red", "green"], alpha=1.0)
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```
