# RVALUE Indicator Example

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume

from mplchart.indicators import TSF, RVALUE
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python

chart = Chart(prices, title=ticker, max_bars=250).plot(
    Candlesticks(),
    TSF(20),
    Volume(),
).plot(
    RVALUE(20), target="below"
).show()
```
