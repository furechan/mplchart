# Stochastic Oscillator Example

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume

from mplchart.indicators import SMA, STOCH
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
Chart(prices, title=ticker, max_bars=250).plot(
    Candlesticks(),
    SMA(50),
    SMA(200),
    Volume(),
).plot(
    STOCH(), target = "below"
).show()
```
