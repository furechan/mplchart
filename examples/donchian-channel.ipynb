# Donchian Channel Example

```python
from mplchart.chart import Chart

from mplchart.samples import sample_prices
from mplchart.primitives import Candlesticks, Volume, SameAxes
from mplchart.indicators import ROC, SMA, EMA, ADX, RSI, MACD, SLOPE, DONCHIAN
```

```python
prices = sample_prices()
```

```python
# Candlesticks with Bollinger Bands

max_bars = 250
indicators = [Candlesticks(), DONCHIAN()]

chart = Chart(title="Donchain Channel", max_bars=max_bars)
chart.plot(prices, indicators)
```
