# Keltner Channel Example

```python
import yfinance as yf

from mplchart.chart import Chart

from mplchart.primitives import Candlesticks, Volume, SameAxes
from mplchart.indicators import ROC, SMA, EMA, ADX, RSI, MACD, SLOPE, KELTNER
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
# Candlesticks with Bollinger Bands

max_bars = 250
indicators = [Candlesticks(), KELTNER()]

Chart(prices, title=ticker, max_bars=max_bars).plot(indicators).show()
```
