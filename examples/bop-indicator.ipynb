# Balance of Power Example

```python
from mplchart.chart import Chart

from mplchart.samples import sample_prices
from mplchart.primitives import Candlesticks, Volume, SameAxes
from mplchart.indicators import ROC, SMA, EMA, ADX, RSI, MACD, SLOPE, HMA, BOP
```

```python
prices = sample_prices()
prices.info()
```

```python
# Candlesticks with Balance of Power

Chart(prices, title="Balance of Power", max_bars=250).plot(
    Candlesticks(), SMA(20), SMA(50)
).plot(
    BOP(14), target="below"
).show()
```
