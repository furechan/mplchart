# Chart Primitives Gallery

```python
import yfinance as yf

from mplchart.chart import Chart

from mplchart.indicators import SMA, EMA, ROC, MACD, PPO, SLOPE

from mplchart.primitives import  Candlesticks, OHLC, Price, Volume, SameAxes, LinePlot
```

## Fetching the prices

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

## Candlestick plot

```python
Chart(prices, title=ticker, max_bars=250, figsize=(12, 9)).plot(
    Candlesticks()
).show()
```

## Open High Low Close bars

```python
Chart(prices, title=ticker, max_bars=250, figsize=(12, 9)).plot(
    OHLC()
).show()
```

## Simple price plot

```python
Chart(prices, title=ticker, max_bars=250, figsize=(12, 9)).plot(
    Price()
).show()
```

## Candlesticks and Volume

```python
Chart(prices, title=ticker, max_bars=250, figsize=(12, 9)).plot(
    Candlesticks(), Volume()
).show()
```

## Customizing style with LinePlot

```python
max_bars = 250

indicators = [
    Candlesticks(),
    SMA(20) | LinePlot(style="dashed", color="red", alpha=0.5, width=3)
]

Chart(prices, title=ticker, max_bars=max_bars).plot(indicators).show()
```
