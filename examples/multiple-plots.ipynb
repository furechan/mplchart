# Multiple Asset Plots

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume, Price
from mplchart.indicators import SMA, RSI, MACD
```

## Fetching the prices

```python
ticker = "AAPL"
aapl = yf.Ticker(ticker).history(period="5y")
aapl
```

```python
ticker = "MSFT"
msft = yf.Ticker(ticker).history(period="5y")
msft
```

## Creating the Chart

```python
# Multiple plots without rebase

title = "AAPL vs MSFT"
max_bars = 250

indicators = [
    Candlesticks(),
    Volume(),
    RSI(),
    MACD(),
]

chart = Chart(title=title, max_bars=max_bars)
chart.plot(aapl, indicators)
chart.plot(msft, [Price()], target="main")
```

```python
# Multiple plots with rebase

chart = Chart(title=title, max_bars=max_bars)
chart.plot(aapl, indicators)
chart.plot(msft, [Price()], target="main", rebase=True)
```
