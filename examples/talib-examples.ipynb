# Talib Chart Examples

```python
import pandas as pd
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, OHLC, Volume, LinePlot

from talib.abstract import Function
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices = prices.rename(columns=str.lower).rename_axis(index=str.lower)
prices.info()
```

```python
# Candlesticks chart with SMA, RSI and MACD

title = "Talib Example"
max_bars = 250

indicators = [
    Candlesticks(),
    Function("SMA", 50),
    Function("SMA", 200),
    Function("RSI"),
    Function("MACD"),
]

chart = Chart(title=title, max_bars=max_bars)
chart.plot(prices, indicators)
```

```python
# Bollinger Bands

title = "Bollinger Bands"
max_bars = 150

indicators = [
    Candlesticks(),
    Function("BBANDS", 20),
]

chart = Chart(title=title, max_bars=max_bars)
chart.plot(prices, indicators)
```

```python
# Stochastic Oscillator

title = "Stochastic Oscillator"
max_bars = 150

indicators = [
    Candlesticks(),
    Function("STOCH"),
]

chart = Chart(title=title, max_bars=max_bars)
chart.plot(prices, indicators)
```

```python
# Parabolic Stop and Reverse

title = "Parabolic Stop and Reverse"
max_bars = 150

indicators = [
    Candlesticks(),
    Function("SAR") | LinePlot(style="marker"),
]

chart = Chart(title=title, max_bars=max_bars)
chart.plot(prices, indicators)
```
