# ZigZag Example

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, ZigZag
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
chart = Chart(prices, title=ticker, max_bars=250).plot(
    Candlesticks(),
    ZigZag()
).show()
```
