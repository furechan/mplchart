# Hull Moving Average Example

```python
import yfinance as yf

from mplchart.chart import Chart

from mplchart.primitives import Candlesticks, Volume, SameAxes
from mplchart.indicators import ROC, SMA, EMA, ADX, RSI, MACD, SLOPE, HMA
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
# Candlesticks with Hull Moving Average

max_bars = 250
indicators = [Candlesticks(), HMA(20), ROC(5) @ HMA(20)]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```
