# Hull Moving Average Example

```python
from mplchart.chart import Chart

from mplchart.samples import sample_prices
from mplchart.primitives import Candlesticks, Volume
from mplchart.indicators import ROC, SMA, EMA, ADX, RSI, MACD, SLOPE, HMA
```

```python
# Fetch sample prices

prices = sample_prices()
prices.info()
```

```python
Chart(prices, title="HMA Example", max_bars=250).plot(
    Candlesticks(), HMA(20)
).plot(
    ROC(1) @ HMA(20), target="below"
).show()
```
