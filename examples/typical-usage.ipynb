# Typical Usage

Minimal end-to-end example: load sample prices, plot candlesticks with SMA/RSI/MACD.
Uses the **pandas** backend — for polars, see [`chart-expressions.ipynb`](chart-expressions.ipynb).

```python
from mplchart.chart import Chart
from mplchart.samples import sample_prices
from mplchart.primitives import Candlesticks, Volume, Pane, LinePlot
from mplchart.indicators import SMA, RSI, MACD
```

```python
prices = sample_prices()
```

```python
(Chart(prices, title="Example Chart", max_bars=250)
    .plot(
        Candlesticks(), Volume(), SMA(50), SMA(200),
        Pane("above", yticks=(30, 50, 70)),
        RSI() | LinePlot(overbought=70, oversold=30),
        Pane("below"),
        MACD(),
    )
    .show()
)
```
