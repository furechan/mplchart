# CURVE Indicator Example

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume

from mplchart.indicators import CURVE, QSF
from mplchart.samples import sample_prices
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices = sample_prices()
prices.info()
```

```python
CURVE(20) @ prices.close
```

```python

Chart(prices, title=ticker, max_bars=250).plot(
    Candlesticks(),
    QSF(20),
    Volume(),
).plot(
    CURVE(20), target="below"
).show()
```
