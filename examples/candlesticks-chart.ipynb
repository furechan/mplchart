# Candlesticks Chart Example

```python

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, LinePlot
from mplchart.indicators import SMA, EMA
from mplchart.samples import sample_prices
```

```python
prices = sample_prices()
prices.info()
```

```python
# Candlesticks with Moving Averages

max_bars = 250

indicators = [
    Candlesticks(alpha=0.5),
    SMA(20) | LinePlot(width=2, alpha=0.5),
    SMA(50) | LinePlot(width=2, alpha=0.5)
]

chart = Chart(title="Candlesticks Chart", max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```

```python
# Candlesticks with colored candles

max_bars = 250

indicators = [ Candlesticks(use_bars=True, colorup="green", colordn="red") ]

chart = Chart(title="Candlesticks Chart", max_bars=max_bars)
chart.plot(prices, indicators)
chart.show()
```
