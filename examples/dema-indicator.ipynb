# Double Exponential Moving Average Example

```python
from mplchart.chart import Chart

from mplchart.samples import sample_prices
from mplchart.primitives import Candlesticks, Volume, SameAxes
from mplchart.indicators import ROC, SMA, EMA, ADX, RSI, MACD, SLOPE, DEMA
```

```python
# Fetch sample prices

prices = sample_prices()
prices.info()
```

```python
# Create chart with some indicators

Chart(prices, title="DEMA Example", max_bars=250).plot(
    Candlesticks(), DEMA(20)
).plot(
    ROC(1) @ DEMA(20), target="below"
).show()
```

```python
DEMA(20) @ prices
```
