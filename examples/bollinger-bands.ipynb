# Bollinger Bands Example

```python
from mplchart.chart import Chart

from mplchart.samples import sample_prices
from mplchart.primitives import Candlesticks, Volume, SameAxes
from mplchart.indicators import ROC, SMA, EMA, ADX, RSI, MACD, SLOPE, BBANDS, BBP, BBW
```

```python
prices = sample_prices()
prices.info()
```

```python
# Candlesticks with Bollinger Bands

Chart(prices, title="Bollinger Bands", max_bars=250).plot(
    Candlesticks(), Volume(), BBANDS(), 
).plot(
    BBP(), target="above"
).plot(
    BBW(), target="below"
).show()
```
