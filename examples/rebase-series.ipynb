# Rebase Series Example

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume, Price
from mplchart.indicators import SMA, RSI, MACD
```

```python
ticker = "AAPL"
aapl = yf.Ticker(ticker).history(period="5y")
aapl = aapl.rename(columns=str.lower).rename_axis(index=str.lower)
aapl
```

```python
ticker = "MSFT"
msft = yf.Ticker(ticker).history(period="5y")
msft = msft.rename(columns=str.lower).rename_axis(index=str.lower)
msft
```

```python
from mplchart.pandas import merge_prices

prices = merge_prices(aapl, msft=msft.close, rebase=True)
prices
```

```python
# Plot with merged series 

title = "AAPL vs MSFT"
max_bars = 250

indicators = [
    Candlesticks(),
    Price("msft"),
    Volume(),
    RSI(),
    MACD(),
]

chart = Chart(title=title, max_bars=max_bars)
chart.plot(prices, indicators)
```
