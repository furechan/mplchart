# Stripes Primitive Example

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import Candlesticks, Volume, Stripes

from mplchart.indicators import SMA, RSI, MACD
```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
# Stripes primitive will mark areas where the macdhist is positive
# By default the primitive is applied to the last indicator that was plotted

max_bars = 250

indicators = [
    Candlesticks(),
    SMA(50),
    SMA(200),
    Volume(),
    RSI(),
    MACD(),
    Stripes("macdhist", color="green", alpha=0.2)
]

chart = Chart(title=ticker, max_bars=max_bars)
chart.plot(prices, indicators)
```
