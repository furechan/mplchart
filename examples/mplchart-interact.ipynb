```python
import matplotlib.pyplot as plt

from mplchart.chart import Chart
from mplchart.samples import sample_prices
from mplchart.primitives import Volume, Candlesticks

from mplchart.indicators import SMA, RSI, MACD, PPO, SLOPE

from ipywidgets import widgets, interact

from IPython.display import clear_output

from matplotlib_inline.backend_inline import flush_figures, set_matplotlib_formats

set_matplotlib_formats("svg")
```

```python
plt.isinteractive()
```

```python
ticker = widgets.Text(
    value="AAPL",
    placeholder="Enter ticker ...",
    description="Ticker:",
    continuous_update=False,
)
```

```python
indicators = [Candlesticks(), SMA(50), SMA(200), Volume(), RSI(), PPO(), SLOPE()]


@interact(ticker=ticker, max_bars=250)
def draw_chart(ticker, max_bars=250):
    if not ticker:
        return "Nothing to show!"
    data = sample_prices()
    chart = Chart(title=ticker, max_bars=max_bars)
    chart.plot(data, indicators)
    plt.show()
```

```python
tickers = "AAPL,MSCI,AMZN".split(",")
tickers = iter(tickers)


@interact.options(manual=True, manual_name="Next Chart!")
def next_chart():
    ticker = next(tickers, None)
    return draw_chart(ticker)
```

```python
output = widgets.Output()


def update_chart(*args):
    symbol = ticker.value
    with output:
        clear_output()
        draw_chart(symbol)
        flush_figures()


button = widgets.Button(description="Plot")
button.on_click(update_chart)
header = widgets.HBox([ticker, button])
vbox = widgets.VBox([header, output])
vbox
```
