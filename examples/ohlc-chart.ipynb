# Open Hig Low Close Bar Chart Example

```python
import yfinance as yf

from mplchart.chart import Chart
from mplchart.primitives import OHLC
from mplchart.indicators import SMA, EMA

```

```python
ticker = "AAPL"
prices = yf.Ticker(ticker).history(period="5y")
prices.info()
```

```python
# Open High Low Close Bar Chart with Moving Average

Chart(prices, title=ticker, max_bars=250).plot(
    OHLC(),
    SMA(20),
    SMA(50)
).show()
```

```python
# open High Low Close Bar Chart with custom colors

Chart(prices, title=ticker, max_bars=250).plot(
    OHLC(colorup="g", colordn="r")
).show()
```
